```python
# Smart Career Guidance Chatbot

# Mapping of individual skills to potential careers
skill_to_career = {
    "python": ["Data Analyst", "Backend Developer", "Machine Learning Engineer", "Data Scientist"],
    "sql": ["Data Analyst", "Database Administrator"],
    "machine learning": ["Machine Learning Engineer", "Data Scientist"],
    "communication": ["HR", "Customer Service", "Content Writing"],
    "english": ["HR", "Content Writing", "Customer Service"],
    "design": ["Graphic Designer", "UI/UX Designer"],
    "creativity": ["Graphic Designer", "UI/UX Designer"],
}
```

```python
extra_advice = {
    "hello": "Hello! Tell me your skills and I will suggest a career.",
    "hi": "Hello! Tell me your skills and I will suggest a career.",
    "how to learn python": "Start with basics like variables and loops, then do small projects.",
    "cv": "Keep your CV simple, highlight key projects, and use action verbs."
}
```

```python
def get_advice(msg):
    msg_lower = msg.lower()

   # Quick responses for greetings or questions
    for key, advice in extra_advice.items():
        if key in msg_lower:
            return advice

    # Collect careers for all mentioned skills
    suggested_careers = set()
    for skill, careers in skill_to_career.items():
        if skill in msg_lower:
            suggested_careers.update(careers)

    if suggested_careers:
        return "Based on your skills, you could consider: " + ", ".join(suggested_careers)

    return "Tell me your skills (e.g., 'I know Python, SQL')."
```

```python
# --- Chatbot Loop ---
print("Smart Career Guidance Chatbot (type 'exit' to quit)")

while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("Bot: Goodbye! Good luck with your career ðŸ˜Š")
        break
    print("Bot:", get_advice(user_input))
```
